```python
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn import datasets
from scipy.spatial.distance import cdist
import operator
```

```python
def KMeans(k,data):
    
    centroids = random.sample(data,k)
    centroids = np.matrix(centroids)
    flag=1
    while flag==1:
        
        distance_matrix = cdist(data,centroids,'euclidean')
        labels = []

        for i in range(len(distance_matrix)):
            index = np.argmin(distance_matrix[i],axis=0)
            #print(distance_matrix[i],index)
            labels = np.append(labels,index)

        unique_labels = list(set(labels))

        new_centroids = []

        for i in range(len(unique_labels)):
            data_temp = []
            for j in range(len(data)):
                if labels[j] == unique_labels[i]:
                    data_temp = np.append(data_temp,data[j])


            data_temp = np.reshape(data_temp,(-4,4))
            mean = np.mean(data_temp,axis=0)
            new_centroids = np.append(new_centroids,mean)
            new_centroids = np.reshape(new_centroids,(-4,4))
        #    print(new_centroids)
        #print(new_centroids)
        mag = np.linalg.norm(cdist(new_centroids,centroids,'euclidean'))
        if  mag<= 1:
            #print(new_centroids)
            print(labels)
            flag=0
        else:
            #print(cdist(new_centroids,centroids,'euclidean'))
            print(labels)
            centroids = new_centroids
    
    
```

```python
dataset = datasets.load_iris()

data = list(dataset.data)
label = dataset.target

KMeans(3,data)
#print(label)
#random.sample(list(data),3)
```
