```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
uni_array = np.random.uniform(0,1,2000)            #Generate Random Samples
samples = []

for i in range(len(uni_array)):
    if uni_array[i]<0.6:
        samples.append(np.random.normal(5,10))         #Distribution using N(5,10)
    else:
        samples.append(np.random.normal(60,15))        #Distribution using N(60,15)
        
plt.rcParams["patch.force_edgecolor"] = True
count,bins,ignored = plt.hist(samples,30,normed = True,linewidth = 1.2)
plt.xlabel('random value')
plt.ylabel('pdf')
plt.title('Histogram of Random Samples')
plt.show()
```

```python
def kernel(bins,samples,h):               #Generate Gaussian Kernel using e^(-(x-xi)^2/h)/sqrt(2*pi)
    pdf = []
    for i in range(len(bins)):
        temp = []
        for j in range(len(samples)):
            x = np.power((bins[i]-samples[j])/h,2)
            x = np.exp(-x/2)
            temp = np.append(temp,x/np.sqrt(2*np.pi))          
            
        actual = np.average(temp)
        pdf = np.append(pdf,actual)
    
    
    plt.plot(bins,pdf,color = 'r')
    plt.xlabel('pdf')
    plt.ylabel('kernel value')
    plt.show()
    return pdf

#Silverman's thumb rule
variance = np.var(samples)
std_d = np.sqrt(variance)
var = np.power(std_d,5)
x = (4*var)/(3*len(samples))
h = np.power(x,1/5)
    
```

```python
pdf0 = kernel(bins,samples,h)                     #using h calculated as per Silverman's Thumb rule
```

```python
count,bins,ignored = plt.hist(samples,30,normed = True,linewidth = 1.2)
plt.plot(bins,pdf0,color = 'r')
plt.xlabel('Sample values')
plt.ylabel('pdf')
plt.title('Histogram vs Kernel')
plt.show()
```

```python
pdf1 = kernel(bins,samples,h*0.2)              #using h*0.2
```

```python
count,bins,ignored = plt.hist(samples,30,normed = True,linewidth = 1.2)
plt.plot(bins,pdf1,color = 'r')
plt.xlabel('Sample values')
plt.ylabel('pdf')
plt.title('Histogram vs Kernel')
plt.show()
```
